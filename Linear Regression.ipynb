```python
import pandas as pd
df=pd.read_csv(r"C:\Users\pk\Downloads\student.csv")
print(df)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
X=df[['Hours']]
y=df['Score']
model.fit(X,y)
hours=float(input("Enter the Hours ="))
predict=model.predict(pd.DataFrame({'Hours':[hours]}))
print(predict)
```

```python

# Plotting
import matplotlib as plt
from matplotlib import pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Actual Scores')           # Actual points
plt.plot(X, model.predict(X), color='red', label='Regression Line') # Line
plt.scatter(hours, predict, color='green', s=100, label='Your Prediction') # Predicted point
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.title("Linear Regression: Hours vs Score")
plt.legend()
plt.show()
```
